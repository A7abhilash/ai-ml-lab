```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.array([[3,4],[5,6]])
b = np.array([12,2])
x = np.linalg.solve(a,b)
```

```python
print(f"A:\n {a}\nB: {b}\nX: {x}\n")
print(f"In this case, Ax=b is {np.allclose(np.dot(a,x),b)}")
```

```python
a = np.array([[3, 4], [5, 6]])
b = np.array([12,2])
aInv = np.linalg.inv(a)
x = np.matmul(aInv, b)
```
